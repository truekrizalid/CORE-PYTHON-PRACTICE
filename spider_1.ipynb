```python
import re
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import os
'''
html = urlopen("https://www.meituri.com/a/28869/").read().decode('utf-8')

soup = BeautifulSoup(html, features='html.parser')

img_links = soup.find_all("a", {"href": re.compile('.*?\.html')})
for link in img_links:
    print(link['href'])
'''    
    
def  getImgPage(urls):
    html = urlopen(urls).read().decode('utf-8')
    soup = BeautifulSoup(html,features = 'html.parser')
    img_links = soup.find_all("a", {"href": re.compile('.*?\.html')})
    URL = []
    for link in img_links:
        URL.append(link['href'])
    del URL[-1]
    return URL
    




def getImgUrl(urls):
    html = urlopen(urls).read().decode('utf-8')
    soup = BeautifulSoup(html,features = 'html.parser')
    img_links = soup.find_all("img", {"src": re.compile('.*?\.jpg')})
    URL = []
    for link in img_links:
        if link['src'][-6:] != "/0.jpg" :
            URL.append(link['src'])
    
    return URL
    
MT ='https://www.meituri.com/a/5274/'
pages = getImgPage(MT)

imgs = []
for page in pages :
    picUrls = getImgUrl(page)
    for picUrl in picUrls :
        imgs.append(picUrl)
        
print(imgs)


if not os.path.exists('D:\\setup\\image\\gxy\\'):
    os.makedirs('D:\\setup\\image\\gxy\\')
i = 0
for img in imgs:
    pic = requests.get(img,timeout=5) #超时异常判断 5秒超时
    
    
    file_name = 'D:\\setup\\image\\gxy\\'+str(i)+".jpg" #拼接图片名
    print(file_name)
 #将图片存入本地
    fp = open(file_name,'wb')
    fp.write(pic.content) #写入图片
    fp.close()
    i+=1














```
